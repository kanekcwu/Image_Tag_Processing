### Find word similarity with <i>model.wv.most_similar</i>

```python
photo_keyword_df['x']
```

```python
labels = ['synonyms', 'percentage']
result = pd.DataFrame()

for i in photo_keyword_df['x']:
    try:
        temp = model.wv.most_similar([i],topn=10)
        temp_df = pd.DataFrame.from_records(temp, columns=labels)
        temp_df['Keyword'] = i
        result = result.append(temp_df, ignore_index=True)
    except:
        pass

result
```

```python
result.to_csv("synonyms_result.csv")
```
